```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Load the Excel file
xlsx = pd.ExcelFile("C:/Users/parsh/OneDrive/Desktop/Stock_Data.xlsx")

# Loop through each sheet
for sheet_name in xlsx.sheet_names:
    print(f"\n--- EDA for {sheet_name} ---")
    
    # Read and clean data
    df = xlsx.parse(sheet_name)
    df.columns = df.columns.str.strip()

    if 'Date' not in df.columns or 'Adj Close' not in df.columns:
        print(f"Skipping {sheet_name} - required columns not found.")
        continue

    df = df[df['Date'].notna()]
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date', 'Adj Close'])
    df.set_index('Date', inplace=True)
    ts = df['Adj Close'].asfreq('B').fillna(method='ffill')

    # Head and missing values
    print("Head of the dataset:")
    print(df.head())
    print("\nMissing values:")
    print(df.isnull().sum())

    # Time series plot
    plt.figure(figsize=(12, 5))
    plt.plot(ts)
    plt.title(f"{sheet_name} - Adjusted Close Price")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Outlier detection
    Q1 = ts.quantile(0.25)
    Q3 = ts.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ts[(ts < lower) | (ts > upper)]
    print(f"Number of outliers: {len(outliers)}")

    # Boxplot
    plt.figure(figsize=(8, 3))
    sns.boxplot(x=ts)
    plt.title(f"{sheet_name} - Boxplot of Adj Close")
    plt.tight_layout()
    plt.show()

    # ADF test (original)
    result = adfuller(ts)
    print("\nADF Test on Original Series")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.4f}")
    if result[1] < 0.05:
        print("The series is likely stationary.")
    else:
        print("The series is likely non-stationary.")

    # Differencing and ADF again
    ts_diff = ts.diff().dropna()
    result_diff = adfuller(ts_diff)
    print("\nADF Test on First Differenced Series")
    print(f"ADF Statistic: {result_diff[0]:.4f}")
    print(f"p-value: {result_diff[1]:.4f}")
    if result_diff[1] < 0.05:
        print("The differenced series is likely stationary.")
    else:
        print("The differenced series is still non-stationary.")

    # Plot differenced series
    plt.figure(figsize=(12, 5))
    plt.plot(ts_diff)
    plt.title(f"{sheet_name} - First Order Differencing")
    plt.xlabel("Date")
    plt.ylabel(" Adj Close")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import warnings
warnings.filterwarnings("ignore")

# Load Excel file with 5 sheets
xlsx = pd.ExcelFile("C:/Users/parsh/OneDrive/Desktop/Stock_Data.xlsx")

for sheet_name in xlsx.sheet_names:
    print(f"\n--- Decomposition for {sheet_name} ---")
    
    df = xlsx.parse(sheet_name)
    df.columns = df.columns.str.strip()

    if 'Date' not in df.columns or 'Adj Close' not in df.columns:
        print(f"Skipping {sheet_name} - required columns not found.")
        continue

    df = df[df['Date'].notna()]
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date', 'Adj Close'])
    df.set_index('Date', inplace=True)

    ts = df['Adj Close'].asfreq('B').fillna(method='ffill')

    # Additive decomposition
    try:
        additive = seasonal_decompose(ts, model='additive', period=30)

        plt.figure(figsize=(12, 8))
        plt.subplot(411)
        plt.plot(additive.observed)
        plt.title(f"{sheet_name} - Additive Observed")

        plt.subplot(412)
        plt.plot(additive.trend)
        plt.title("Trend")

        plt.subplot(413)
        plt.plot(additive.seasonal)
        plt.title("Seasonality")

        plt.subplot(414)
        plt.plot(additive.resid)
        plt.title("Residuals")

        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Additive decomposition failed for {sheet_name}: {e}")

    # Multiplicative decomposition
    try:
        multiplicative = seasonal_decompose(ts, model='multiplicative', period=30)

        plt.figure(figsize=(12, 8))
        plt.subplot(411)
        plt.plot(multiplicative.observed)
        plt.title(f"{sheet_name} - Multiplicative Observed")

        plt.subplot(412)
        plt.plot(multiplicative.trend)
        plt.title("Trend")

        plt.subplot(413)
        plt.plot(multiplicative.seasonal)
        plt.title("Seasonality")

        plt.subplot(414)
        plt.plot(multiplicative.resid)
        plt.title("Residuals")

        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Multiplicative decomposition failed for {sheet_name}: {e}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
from datetime import timedelta
import warnings
warnings.filterwarnings("ignore")

# Load Excel
xlsx = pd.ExcelFile("C:/Users/parsh/OneDrive/Desktop/Stock_Data.xlsx")
future_end = pd.to_datetime("2025-12-31")

def evaluate(true, pred):
    rmse = np.sqrt(mean_squared_error(true, pred))
    mae = mean_absolute_error(true, pred)
    mape = np.mean(np.abs((true - pred) / true)) * 100
    mse = mean_squared_error(true, pred)
    return rmse, mae, mape, mse

for sheet_name in xlsx.sheet_names:
    print(f"\n--- Forecasting for {sheet_name} ---")
    
    df = xlsx.parse(sheet_name)
    df.columns = df.columns.str.strip()

    if 'Date' not in df.columns or 'Adj Close' not in df.columns:
        print(f"Skipping {sheet_name} - required columns not found.")
        continue

    df = df[df['Date'].notna()]
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date', 'Adj Close'])
    df.set_index('Date', inplace=True)

    ts = df['Adj Close'].asfreq('B').fillna(method='ffill')

    if len(ts) < 50:
        print(f"Skipping {sheet_name} - not enough data.")
        continue

    train_size = int(len(ts) * 0.8)
    train, test = ts[:train_size], ts[train_size:]

    try:
        arima_model = ARIMA(train, order=(5, 1, 2)).fit()
        arima_pred = arima_model.forecast(steps=len(test))
        arima_scores = evaluate(test, arima_pred)
        arima_extended = arima_model.forecast(
            steps=len(pd.date_range(start=ts.index[-1] + timedelta(days=1), end=future_end, freq='B'))
        )
    except Exception as e:
        print(f"ARIMA failed for {sheet_name}: {e}")
        arima_pred, arima_extended = None, None
        arima_scores = [np.nan] * 4

    try:
        ets_model = ExponentialSmoothing(train, trend='add', seasonal=None).fit()
        ets_pred = ets_model.forecast(steps=len(test))
        ets_scores = evaluate(test, ets_pred)
        ets_extended = ets_model.forecast(
            steps=len(pd.date_range(start=ts.index[-1] + timedelta(days=1), end=future_end, freq='B'))
        )
    except Exception as e:
        print(f"ETS failed for {sheet_name}: {e}")
        ets_pred, ets_extended = None, None
        ets_scores = [np.nan] * 4

    try:
        df_prophet = ts.reset_index().rename(columns={'Date': 'ds', 'Adj Close': 'y'})
        prophet_model = Prophet(daily_seasonality=True)
        prophet_model.fit(df_prophet.iloc[:train_size])
        future = prophet_model.make_future_dataframe(periods=len(test), freq='B')
        forecast = prophet_model.predict(future)
        prophet_pred = forecast['yhat'].iloc[-len(test):].values
        prophet_scores = evaluate(test.values, prophet_pred)

        full_prophet = Prophet(daily_seasonality=True)
        full_prophet.fit(df_prophet)
        steps_extended = len(pd.date_range(start=ts.index[-1] + timedelta(days=1), end=future_end, freq='B'))
        future_ext = full_prophet.make_future_dataframe(periods=steps_extended, freq='B')
        forecast_ext = full_prophet.predict(future_ext)
        prophet_extended = forecast_ext['yhat'].iloc[-steps_extended:].values
        prophet_extended_index = forecast_ext['ds'].iloc[-steps_extended:]
    except Exception as e:
        print(f"Prophet failed for {sheet_name}: {e}")
        prophet_pred, prophet_extended, prophet_extended_index = None, None, None
        prophet_scores = [np.nan] * 4

    # Plot short-term forecast
    plt.figure(figsize=(12, 5))
    plt.plot(ts.index, ts, label='Actual')
    if arima_pred is not None:
        plt.plot(test.index, arima_pred, label='ARIMA')
    if ets_pred is not None:
        plt.plot(test.index, ets_pred, label='ETS')
    if prophet_pred is not None:
        plt.plot(test.index, prophet_pred, label='Prophet')
    plt.title(f"{sheet_name} - Short-Term Forecast")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot extended forecast
    plt.figure(figsize=(12, 5))
    plt.plot(ts.index, ts, label='Actual')
    if arima_extended is not None:
        ext_idx = pd.date_range(start=ts.index[-1] + timedelta(days=1), periods=len(arima_extended), freq='B')
        plt.plot(ext_idx, arima_extended, label='ARIMA (Extended)')
    if ets_extended is not None:
        ext_idx = pd.date_range(start=ts.index[-1] + timedelta(days=1), periods=len(ets_extended), freq='B')
        plt.plot(ext_idx, ets_extended, label='ETS (Extended)')
    if prophet_extended is not None:
        plt.plot(prophet_extended_index, prophet_extended, label='Prophet (Extended)')
    plt.title(f"{sheet_name} - Forecast Extended to Dec 2025")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid
```

```python
code = '''
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings("ignore")

st.set_page_config(page_title=" Time Series Forecast App", layout="wide")
st.title(" Time Series Forecasting Web App")
st.markdown("Upload an Excel file with time series data. Select a sheet, choose decomposition & model, and view forecasts.")

# File uploader
uploaded_file = st.file_uploader(" Upload an Excel file", type=["xlsx"])

# Forecasting model and decomposition type selection
model_choice = st.selectbox("Choose Forecasting Model", ["ARIMA", "ETS", "Prophet"])
decomp_choice = st.radio("Choose Decomposition Type", ["Additive", "Multiplicative"])

# Evaluation metric function
def evaluate(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    mse = mean_squared_error(y_true, y_pred)
    return rmse, mae, mape, mse

if uploaded_file:
    try:
        xlsx = pd.ExcelFile(uploaded_file)
        sheet = st.selectbox(" Select Sheet", xlsx.sheet_names)
        df = xlsx.parse(sheet)
        df.columns = df.columns.str.strip()

        date_col = "Date" if "Date" in df.columns else df.columns[0]
        value_col = df.columns[1]

        df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
        df = df.dropna(subset=[date_col, value_col])
        df.set_index(date_col, inplace=True)
        ts = df[value_col].asfreq('B').fillna(method='ffill')

        st.subheader(" Raw Time Series")
        st.line_chart(ts)

        # Decomposition
        try:
            result = seasonal_decompose(ts, model=decomp_choice.lower(), period=30)
            st.subheader(" Time Series Decomposition")
            fig, axs = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
            result.observed.plot(ax=axs[0], title='Observed')
            result.trend.plot(ax=axs[1], title='Trend')
            result.seasonal.plot(ax=axs[2], title='Seasonality')
            result.resid.plot(ax=axs[3], title='Residuals')
            st.pyplot(fig)
        except Exception as e:
            st.warning(f"Decomposition failed: {e}")

        # Forecasting
        train_size = int(len(ts) * 0.8)
        train, test = ts[:train_size], ts[train_size:]
        future_dates = pd.date_range(start=test.index[-1] + pd.Timedelta(days=1), periods=60, freq='B')

        st.subheader(" Forecast Plot")

        if model_choice == "ARIMA":
            model = ARIMA(train, order=(5, 1, 2)).fit()
            forecast = model.forecast(steps=len(test))
            future_forecast = model.forecast(steps=60)

        elif model_choice == "ETS":
            model = ExponentialSmoothing(train, trend='add', seasonal=None).fit()
            forecast = model.forecast(steps=len(test))
            future_forecast = model.forecast(steps=60)

        elif model_choice == "Prophet":
            prophet_df = ts.reset_index().rename(columns={ts.index.name: "ds", ts.name: "y"})
            prophet_model = Prophet(daily_seasonality=True)
            prophet_model.fit(prophet_df.iloc[:train_size])
            future = prophet_model.make_future_dataframe(periods=len(test) + 60, freq='B')
            forecast_df = prophet_model.predict(future)
            forecast = forecast_df['yhat'].iloc[train_size:train_size+len(test)].values
            future_forecast = forecast_df['yhat'].iloc[-60:].values

        fig, ax = plt.subplots(figsize=(12, 5))
        ax.plot(ts.index, ts, label="Actual")
        ax.plot(test.index, forecast, label="Forecast")
        ax.plot(future_dates, future_forecast, label="Future Forecast (60 days)", linestyle="--")
        ax.set_title(f"{model_choice} Forecast")
        ax.legend()
        ax.grid(True)
        st.pyplot(fig)

        # Evaluation
        rmse, mae, mape, mse = evaluate(test, forecast)
        st.subheader(" Forecast Accuracy Metrics")
        st.write(f"**RMSE:** {rmse:.4f}")
        st.write(f"**MAE:** {mae:.4f}")
        st.write(f"**MAPE:** {mape:.2f}%")
        st.write(f"**MSE:** {mse:.4f}")

    except Exception as e:
        st.error(f" Something went wrong: {e}")

# Footer in bottom-right corner
st.markdown(
    """
    <style>
        .footer {
            position: fixed;
            bottom: 10px;
            right: 10px;
            font-size: 13px;
            color: gray;
        }
    </style>
    <div class="footer"> Designed by <strong>Parshva</strong></div>
    """,
    unsafe_allow_html=True
)
'''

with open("app.py", "w", encoding="utf-8") as f:
    f.write(code)

```
